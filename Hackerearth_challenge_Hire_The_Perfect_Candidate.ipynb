Function to remove Punctuation

```python
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    temp=text_nopunct.split()
    x=[]
    for j in range(len(temp)):
      if(len(temp[j])==1):
        x.append(temp[j])
        temp[j]=""
      elif(len(x)==0 or temp[j]==' '):
        continue
      else:
        temp.append("".join(x))
        x=[]
    text_nopunct=' '.join(temp)
    return text_nopunct

data['resume_text_clean'] = data['resume'].apply(lambda x: remove_punct(x))

data.head()
```

Function to Tokenize words

```python
def tokenize(text):
    tokens = re.split('\W+', text) #W+ means that either a word character (A-Za-z0-9_) or a dash (-) can go there.
    return tokens

data['resume_text_tokenized'] = data['resume_text_clean'].apply(lambda x: tokenize(x.lower())) 
#We convert to lower as Python is case-sensitive. 

data.head()
```

Function to remove Stopwords

```python
stopword = nltk.corpus.stopwords.words('english')# All English Stopwords
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]# To remove all stopwords
    return text

data['resume_text_nostop'] = data['resume_text_tokenized'].apply(lambda x: remove_stopwords(x))

data.head()
```

Preprocessing Data: Using Stemming

```python
ps = nltk.PorterStemmer()

def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

data['resume_text_stemmed'] = data['resume_text_nostop'].apply(lambda x: stemming(x))

data.head()
```

Preprocessing Data: Using a Lemmatizer

```python
wn = nltk.WordNetLemmatizer()

def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

data['resume_text_lemmatized'] = data['resume_text_nostop'].apply(lambda x: lemmatizing(x))

data.head(10)
```
